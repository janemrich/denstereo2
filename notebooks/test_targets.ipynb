```python
# load test_pbr_left.json

import json
```

```python
targets_json = '/workspace/denstereo-so/datasets/BOP_DATASETS/denstereo/test_bboxes/test_pbr_left.json'
targets_json_out = '/workspace/denstereo-so/datasets/BOP_DATASETS/denstereo/test_bboxes/test_targets_pbr_left.json'

with open(targets_json) as f:
    data = json.load(f)
```

```python
out = []
for scene_im_id in data:
    for bbox_dict in data[scene_im_id]:
        # print(bbox_dict)
        instance = {}
        scene, im = scene_im_id.split('/')
        instance['im_id'] = int(im)
        instance['inst_count'] = 1
        instance['obj_id'] = bbox_dict['obj_id']
        instance['scene_id'] = int(scene)
        out.append(instance)
```

```python
# save out to test_targets.json

with open(targets_json_out, 'w') as f:
    json.dump(out, f)
```
